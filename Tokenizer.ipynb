# tokenizer的基本使用

## tokenizer的初始化

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("uer/roberta-base-finetuned-dianping-chinese")
tokenizer
```

```python
#将模型文件保存到本地
tokenizer.save_pretrained("./roberta_tokenizer")
```

```python
#从本地加载模型
tokenizer = AutoTokenizer.from_pretrained("./roberta_tokenizer")
tokenizer
```

## 分词

```python
#存储句子
sen = "今天天气真好,我想要吃火锅！"
tokens = tokenizer.tokenize(sen)
tokens
```

## 查看词典

```python
tokenizer.vocab
```

## 索引转换

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
ids
```

```python
tokens_2 = tokenizer.convert_ids_to_tokens(ids)
stren = tokenizer.convert_tokens_to_string(tokens_2)
stren
```

```python
#便捷的方法——编码
ids = tokenizer.encode(sen)
ids
#在开头与结尾的位置特殊的序号是标记句子的开头与结尾，所有的模型都有不一样的规则
```

```python
#便捷的方法——解码
strs = tokenizer.decode(ids)
strs
#特殊的开始与结尾的标志CLS与SEP可以选择跳过
strs = tokenizer.decode(ids,skip_special_tokens=True)
strs
```

## 填充与截断

```python
#填充
ids = tokenizer.encode(sen,padding = "max_length",max_length = 20)
ids
```

```python
#截断
ids = tokenizer.encode(sen,max_length=5,truncantion = True)
ids
```

## 屏蔽填充的部分

```python
ids = tokenizer.encode(sen,max_length = 20,padding = "max_length")
ids
```

```python
attention_mask = [1 if idx!=0 else 0 for idx in ids]
token_type_ids = [0]*len(ids)
ids,token_type_ids,attention_mask
```

## 快捷调用方法

```python
#旧版需要调用方法，但是现版需要我们直接使用_call_方法直接调用
inputs = tokenizer(sen , padding = "max_length" , max_length = 30)
inputs
```

## batch方法处理数据

```python
%%time
sens = [
    "宝宝嘟嘟打雷了，",
    "雷雷肚肚打宝宝，",
    "肚肚宝宝打雷了，我要吃外卖"
]
inputs = tokenizer(sens,padding = True,return_tensors="pt")
inputs
```

## fasttokenizer/slowtokenizer

```python
fast_tokenizer =  AutoTokenizer.from_pretrained("./roberta_tokenizer")
#fast使用的是rust语言实现的，比较快，而且会多一点其他的数据预处理的方式
slow_tokenizer = AutoTokenizer.from_pretrained("./roberta_tokenizer",use_fast = False)
#slow是使用python语言实现的，会相对于比较慢一点
```

```python
sen = "小小的我也有大大的dreaming"
inputs = fast_tokenizer(sen,return_offsets_mapping=True)
inputs,inputs.word_ids()
#offsets_mapping告诉我们每一个Token，对应原始文本中的哪一段连续字符
#word_ids 则提供了一份单词级别的地图。它的作用是告诉我们每一个Token，属于原始文本中的第几个“单词”
```

## 加载其他大模型

```python
# Hugging Face的 transformers 库内置了许多标准模型架构（如BERT、GPT等）。然而，很多先进的大语言模型（LLMs）会采用创新的网络结构或特殊的分词逻辑，这些并不在官方标准库中。
# 为了能加载这些模型，Hugging Face允许模型仓库包含自定义的Python代码（如 modeling.py, tokenization.py）。trust_remote_code=True 这个参数，就是你明确授权 transformers 去下载并执行这些自定义代码的信号
from transformers import AutoTokenizer
my_tokenizer = AutoTokenizer.from_pretrained("zai-org/GLM-5.2",trust_remote_code = True)
my_tokenizer
```

```python
input = my_tokenizer(sen,return_tensors="pt")
input
```
